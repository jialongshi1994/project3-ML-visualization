```python
import requests
import json
import pandas as pd
from pprint import pprint
from datetime import datetime
import wbdata 
```

```python
jsonfile = requests.get("https://api.covid19api.com/countries")
country = jsonfile.json()
```

```python
country_name = []
county_ISO2 = []
county_slug = []

for i in range(len(country)):
    country_name.append(country[i]["Country"])
    county_ISO2.append(country[i]["ISO2"])
    county_slug.append(country[i]["Slug"])
```

```python
jsonfile_2 = requests.get("http://country.io/iso3.json")
iso_code = jsonfile_2.json()
```

```python
base_url = "https://api.covid19api.com/total/country/"

covid_df = pd.DataFrame(columns = ["country", "ISO2", "ISO", "slug","start_date", "end_date", "lable_end_date", "active","new_cases_feature","new_cases_label"])

for i in range(len(county_slug)):
    request = requests.get(base_url +  county_slug[i])
    json = request.json()
    
    start = 20
    
    
    for k in range(4):
        
        counter = 20
        
    
    

        for x in range(start, len(json) - counter, counter):

            if len(json) > 100:

                confirmed = json[x]["Confirmed"]
                confirmed_2 = json[x - counter]["Confirmed"]
                confirmed_3 = json[x + counter]["Confirmed"]

                new_cases_feature = confirmed - confirmed_2
                new_cases_label = confirmed_3 - confirmed_2

                active = json[x]["Active"]
                end_date = json[x]["Date"]
                start_date = json[x - counter]["Date"]
                lable_end_date = json[x + counter]["Date"]

                covid_df = covid_df.append({"country" : country_name[i], 
                                            "ISO2": county_ISO2[i],
                                            "ISO" : iso_code[county_ISO2[i]],
                                            "slug": county_slug [i],
                                            "start_date":start_date , 
                                            "end_date": end_date, 
                                            "lable_end_date": lable_end_date, 
                                            "active": active,
                                            "new_cases_feature": new_cases_feature,
                                            "new_cases_label" : new_cases_label }, ignore_index=True)
    
        start = start + 5

   
        
```

```python
covid_df["end_date"] = pd.to_datetime(covid_df["end_date"]).dt.strftime('%Y-%m-%d')
covid_df["start_date"] = pd.to_datetime(covid_df["start_date"]).dt.strftime('%Y-%m-%d')
covid_df["lable_end_date"] = pd.to_datetime(covid_df["lable_end_date"]).dt.strftime('%Y-%m-%d')
```

```python
date = datetime(2019, 1, 1)




Urban_population = wbdata.get_data("SP.URB.TOTL.IN.ZS", country = countires5, data_date = data,freq='Y')
UHC_service_coverage_index = wbdata.get_data("SH.UHC.SRVS.CV.XD", country = countires5, data_date = datetime(2017, 1, 1),freq='Y')
basic_sanitation_services = wbdata.get_data("SH.STA.BASS.ZS", country = countires5, data_date = datetime(2017, 1, 1),freq='Y')
Population_ages_0_to_14 = wbdata.get_data("SP.POP.0014.TO.ZS", country = countires5, data_date = date,freq='Y')
Population_ages_15_to_64 = wbdata.get_data("SP.POP.1564.TO.ZS", country = countires5, data_date = date,freq='Y')
Population_ages_65_and_over = wbdata.get_data("SP.POP.65UP.TO.ZS", country = countires5, data_date = date,freq='Y')

```

```python
multiple_data = {}

for i in range(len(Urban_population)):
    multiple_data.update({Urban_population[i]["countryiso3code"] : {"Urban_population" : Urban_population [i]["value"], 
                                                                 "UHC_service_coverage_index": UHC_service_coverage_index [i]["value"],
                                                                "basic_sanitation_services": basic_sanitation_services [i]["value"],
                                                                 "Population_ages_0_to_14": Population_ages_0_to_14 [i]["value"],
                                                                  "Population_ages_15_to_64": Population_ages_15_to_64 [i]["value"],
                                                                   "Population_ages_65_and_over": Population_ages_65_and_over [i]["value"]}})
```

```python
Urban_population_list = []
UHC_service_coverage_index_list = []
basic_sanitation_services_list = []
Population_ages_0_to_14_list = []
Population_ages_15_to_64_list = []
Population_ages_65_and_over_list = []

for i in range(len(covid_df3)):

    d1 = multiple_data[covid_df3.iloc[i,2]]["Urban_population"]
    d2 = multiple_data[covid_df3.iloc[i,2]]["UHC_service_coverage_index"]
    d3 = multiple_data[covid_df3.iloc[i,2]]["basic_sanitation_services"]
    d4 = multiple_data[covid_df3.iloc[i,2]]["Population_ages_0_to_14"]
    d5 = multiple_data[covid_df3.iloc[i,2]]["Population_ages_15_to_64"]
    d6 = multiple_data[covid_df3.iloc[i,2]]["Population_ages_65_and_over"]
  
    
    Urban_population_list.append(d1)
    UHC_service_coverage_index_list.append(d2)
    basic_sanitation_services_list.append(d3)
    Population_ages_0_to_14_list.append(d4)
    Population_ages_15_to_64_list.append(d5)
    Population_ages_65_and_over_list.append(d6)
```

```python
covid_df3["Urban_population"] = Urban_population_list
covid_df3["UHC_service_coverage_index"] = UHC_service_coverage_index_list
covid_df3["basic_sanitation_services"] = basic_sanitation_services_list
covid_df3["Population_ages_0_to_14"] = Population_ages_0_to_14_list
covid_df3["Population_ages_15_to_64"] = Population_ages_15_to_64_list
covid_df3["Population_ages_65_and_over"] = Population_ages_65_and_over_list
```
